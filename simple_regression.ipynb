## 単回帰

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# 説明変数と目的変数のデータセットを用意する
# 説明変数は height (cm)
# 目的変数は weight (kg)

height = np.array([160, 165, 170, 175, 180, 185, 190, 195])
weight = np.array([53, 54, 60, 73, 78, 80, 79, 88])
height = height.reshape(-1,1) # LinearRegression へ与える説明変数の shape は (データ数,1)

# LinearRegression クラスのインスタンスを作成する
model = LinearRegression()

# モデルを学習させる
model.fit(height, weight)

# 予測値を計算する
predicted_weight = model.predict(height)

# 回帰係数を出力する
print("coefficient:", model.coef_)

# 切片を出力する
print("intercept:", model.intercept_)

# 決定係数を出力する
print("coefficient of determination:", model.score(height, weight))

# 予測値を表示する
print("predicted weights:", predicted_weight)

# 散布図
plt.scatter(height, weight)
plt.xlabel("height")
plt.ylabel("weight")

# 回帰直線
xx = np.linspace(height.min(), height.max(), 100).reshape(-1, 1)
plt.plot(xx[:, 0], model.predict(xx), color='r')
```
